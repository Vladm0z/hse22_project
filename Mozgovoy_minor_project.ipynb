```python
import os
import subprocess
import tempfile
from pathlib import Path
from subprocess import DEVNULL, PIPE
```

```python
ZH_EXECUTABLE = Path("/content/zhunt3")
assert ZH_EXECUTABLE.is_file()

def zhunt(query: str, windowsize: int = 6, minsize: int = 3, maxsize: int = 6):
    query = query.upper()
    assert set(query).issubset({"A", "C", "G", "T", "N"})
    fd, temp = tempfile.mkstemp()
    os.close(fd)
    with open(temp, 'w') as stream:
        stream.write(query)

    subprocess.run(
        [ZH_EXECUTABLE, 
         str(windowsize), str(minsize), str(maxsize), temp],
        check=True, stdout=PIPE, stderr=DEVNULL,
        input=query, encoding='ascii'
    )
    with open(temp + ".Z-SCORE", 'r') as stream:
        df = pd.read_csv(stream,
                         names=['Start', 'End', 'nu-1', 'nu-2', 'nu-3', 
                                'ZH-Score', 'Sequence', 'Conformation'],
                         skiprows=1, sep='\s+')
    os.remove(temp)
    os.remove(temp + ".Z-SCORE")
    return df[['Start', 'End', 'ZH-Score', 'Sequence', 'Conformation']]
```

```python
%%time
list_of_files_for_ZDNA = ["/content/GCA_000582765.1_AKH_PRJEB1535_v1_genomic.fna",
                          "/content/GCA_000982615.1_AKI_PRJEB1539_v1_genomic.fna",
                          "/content/GCA_001235845.1_ASM123584v1_genomic.fna",
                          "/content/GCA_001293395.1_ASM129339v1_genomic.fna",
                          "/content/GCA_001299535.1_ASM129953v1_genomic.fna"]

df_list = []

for ij in list_of_files_for_ZDNA:
  with open(ij, 'r') as stream:
    sequences = list(SeqIO.parse(stream, format='fasta'))
    
  for s in sequences:
   print(f"{s.description} \n\t=> {len(s)}")
  
  seq = sequences[0]
  df = zhunt(str(seq.seq))

  df_list.append(df)
```

```python
def get_total_len(df):
  sum = 0
  for j, row in df.iterrows():
    start = row['Start']
    end = row['End']
    s = int(end) - int(start) + 1
    sum += s
  return sum
```

```python
dfs_coord = []
ind_name = 0
for ft in tqdm(feature_list):
  ft_pd = pd.read_csv(ft, sep='\t')
  f = pd.DataFrame(columns=['name','start','end','n'])
  ind = 0
  for i in range(0, ft_pd.shape[0]):
    row = ft_pd.iloc[i]
    if row['# feature'] == 'gene':
      if row.strand == '+':
        pos = row.start
        strand = '+'
      else:
        pos = row.end
        strand = '-'
    elif row['# feature'] == 'CDS':
      f.loc[ind] = [fna_name[ind_name], pos, pos, fna_name[ind_name] + '_' + str(pos)]
      ind += 1
  dfs_coord.append(f)
  f.to_csv(f'{names[ind_name]}_TSS.bed', sep = '\t', header=False,index=False)
  ind_name += 1
```

```python
score = []
scores = []
for i in tqdm(range(0, clust_fin.shape[0])):
  row = data.loc[i]
  genes = [row['GCA_000582765.faa'],
           row['GCA_000982615.faa'],
           row['GCA_001235845.faa'],
           row['GCA_001293395.faa'],
           row['GCA_001299535.faa']]
  sum_inter = 0
  temp_score = []
  for j in range(5):
    f = pd.DataFrame(columns=['name', 'start', 'end', 'n'])
    table = pd.read_csv(feature_list[j], sep = '\t')
    row_gene = table.loc[table[table['product_accession'] == genes[j]].index - 1]
    if row_gene['strand'].values[0] == '+':
      pos = row_gene['start'].values[0]
    else:
      pos = row_gene['end'].values[0]
    f.loc[0] = [fna_name[j], pos, pos, fna_name[j] + '_' + str(pos)]
    f.to_csv('temp.bed', sep = '\t', header=False, index=False)
    len_name = names[j] + '_len.bed'
    ! bedtools slop -i temp.bed -g '{len_name}' -b 300 > temp_slop.bed
    ! intersectBed -a '{final[j]}' -b temp_slop.bed -wb > temp_intersec.bed
    lines = ! wc -l temp_intersec.bed
    if int(lines[0].split(' ')[0]) == 0:
      continue
    temp_score.append(round(pd.read_csv('temp_intersec.bed', sep='\t', header=None).loc[:, 4].sum(), 4))
    inter_num = ! wc -l temp_intersec.bed
    sum_inter += int(inter_num[0].split(' ')[0])
  scores.append(temp_score)
  score.append(round(np.mean(temp_score), 4))
```

```python
clust_final = clust_fin.copy()
clust_final.loc[:, 'zh-score'] = scores
clust_final.loc[:, 'mean zh-score'] = score
clust_final.reset_index(inplace=True)
clust_final
```

```python
for i in tqdm(range(clust_final.shape[0])):
  output = []
  row = clust_final.loc[i]
  for j in range(5):
    prot = clust_final[clust_final.columns[j+4]][i]
    cl_fin = open(clust_final.columns[j+4], "r")
    flag = False
    while True:
      line = cl_fin.readline()
      if not line:
        break
      if '>' in line.strip():
        flag = False
      if flag:
        output.append(line.strip())
      if prot in line.strip():
        output.append(line.strip())
        flag = True
    cl_fin.close
  f = open( ('clust' + str(i) + '.txt'), 'w' )
  f.write('\n'.join(output))
  f.close()
```
